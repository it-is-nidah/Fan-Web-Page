```python
import numpy as np

cat_img = cv2.imread('cat.jpg',0)
equ = cv2.equalizeHist(cat_img)
res = np.hstack((cat_img,equ))
cv2.imwrite('res.png',res)

hist_full = cv2.calcHist([equ],[0],None,[256],[0,256])
plt.plot(hist_full)

cat = cv2.imread('cat.jpg', cv2.IMREAD_GRAYSCALE)

eq = cv2.equalizeHist(cat)

plt.figure(figsize=(10, 5))
plt.subplot(121)
plt.imshow(cat, cmap='gray')
plt.title('Original grayscale Image')
plt.axis('off')
plt.subplot(122)
plt.imshow(eq, cmap='gray')
plt.title('Equalized Image')
plt.axis('off')
plt.show()
```

```python

img = cv2.imread('pic3.jpg', cv2.IMREAD_GRAYSCALE)

# Applying Prewitt edge detector
kernelx = np.array([[1, 1, 1], [0, 0, 0], [-1, -1, -1]])
kernely = np.array([[-1, 0, 1], [-1, 0, 1], [-1, 0, 1]])
prewittx = cv2.filter2D(img, -1, kernelx)
prewitty = cv2.filter2D(img, -1, kernely)
prewitt = cv2.addWeighted(prewittx, 0.5, prewitty, 0.5, 0)

# Applying Sobel edge detector
sobelx = cv2.Sobel(img, cv2.CV_64F, 1, 0, ksize=5)
sobely = cv2.Sobel(img, cv2.CV_64F, 0, 1, ksize=5)
sobel = cv2.addWeighted(sobelx, 0.5, sobely, 0.5, 0)

print('Original')
cv2_imshow( img)
print('Prewitt')
cv2_imshow(prewitt)
print('Sobel')
cv2_imshow(sobel)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img = cv2.imread('pic4.jpg', cv2.IMREAD_GRAYSCALE)

# Apply LOG edge detector
gaussian = cv2.GaussianBlur(img, (5, 5), 0)
laplacian = cv2.Laplacian(gaussian, cv2.CV_64F)

print('original image')
cv2_imshow(img)
print('LOG image')
cv2_imshow(laplacian)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
